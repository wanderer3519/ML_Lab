```python
def f(x):
    return x

def df(x):
    return 1

DELTA = 0.1

def gradientDescent(x, x0, alpha):
    x = x0
    check = 10e10

    beta = 0

    while check > 10e-4:
        obj = f(x)
        beta = 0.9
        alpha = 10

        temp = x - alpha * df(x)
    
    while f(temp) > f(x) - DELTA:
        alpha *= beta
        temp = x - alpha * df(x)

        if(alpha < 10e-5):
            temp = x
            break

    x = temp
    check = obj - f(x)
    return x
```
