```python
import scipy as sp
import numpy as np
```

```python
# 1(1)a Solve a system of linear equations
input_array = np.array([[1, 5, 10, 7], [2, 12, 7, 1], [10, 8, 3, 2], [5, 4, 7, 5]])
constants = np.array([10, 18, 20 ,30])

solution = sp.linalg.solve(input_array, constants)
solution
```

```python
# 1(1)b Verify solution
s = np.matmul(input_array, solution)
s
```

```python
# 1(1)c determinant of sq matrix
input_det = sp.linalg.det(input_array)
input_det
```

```python
# 1(1)c trace of sq matrix
tr = np.trace(input_array)
tr
```

```python
# 1(1)d Eigen values and eigen vectors
evals, right_eig = sp.linalg.eig(input_array)

evals, right_eig
```

```python
# 1(1)e check if input is symmetric
sym = sp.linalg.issymmetric(input_array)
sym
```

```python
# Modify this again
# 1(2)a Convex hull
rng = np.random.default_rng()
points = rng.random((8, 2))
hull = sp.spatial.ConvexHull(points)

# 1(2)b vizualize using matplotlib and 1(2)c
plt.plot(points[:,0], points[:,1], 'o')
for simplex in hull.simplices:
    plt.plot(points[simplex, 0], points[simplex, 1], 'k-')

# 1(2) d fill area within polygon
plt.fill(points[hull.vertices,0], points[hull.vertices,1], 'k', alpha = 0.3)
plt.show()
```

```python
# 2(3)c sort dataframe
df = df.sort_values(by = ['age', 'visits'], ascending = [False, True])
df
```

```python
# 2(3)d Drop duplicates and show dataframe

x = df.drop_duplicates()
df = x
df
```

```python
def yes(values):
    c = 0
    for data in values:
        if(data == 'yes'):
            c += 1
    return c

def no(values):
    c = 0
    for data in values:
        if(data == 'no'):
            c += 1
    return c
       
pivot_df = df.pivot_table(values = 'priority', index = 'birds', aggfunc = [yes, no])
pivot_df.plot(kind='bar', figsize=(10, 6))
plt.title('Frequency of priority vs Type of bird')
plt.xlabel('Type of bird')
plt.ylabel('Frequency')
plt.xticks(rotation=0)  # Rotate x labels for better readability
plt.legend(title='Statistics')
plt.show()
```
