```python
import numpy as np
import pandas as pd
```

```python
x = np.random.uniform(low = 0, high = 5, size = 100)
e = np.random.rand(100)
a = np.random.rand()
b = np.random.rand()

y = np.array([x[i] ** 2 + a * x[i] + b + e[i] for i in range(len(x))])
```

```python
X = np.column_stack([np.ones(100), x])
X
```

```python
beta = np.linalg.inv(X.T @ X) @ X.T @ y
```

```python
beta
copyx = x.copy()
```

```python
from sklearn.metrics import mean_squared_error as mse
y_pred = X @ beta

mse(y_true = y, y_pred = y_pred)
```

```python
from matplotlib import pyplot as plt

plt.title("Plot of data")
plt.xlabel("Feature")
plt.ylabel("Value")
plt.scatter(x, y, label = "Original Data")
plt.scatter(x, y_pred, label = "Predicted Data")
plt.legend()
```

```python
x = copyx
x = x.reshape(-1, 1)
x
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(x, y)
```

```python
y_pred = reg.predict(x)
# y_pred
reg.coef_
```
