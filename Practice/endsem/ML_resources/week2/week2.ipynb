```python
data = {
    'birds': ['Cranes', 'Cranes', 'plovers', 'spoonbills', 'spoonbills', 'Cranes',
            'plovers', 'Cranes', 'spoonbills', 'spoonbills', 'Cranes', 'Cranes', 'plovers',
            'plovers', 'spoonbills', 'spoonbills', 'Cranes', 'plovers', 'spoonbills', 'Cranes',
            'Cranes', 'plovers', 'spoonbills', 'Cranes'],
    'age': [3.5, 4, 1.5, np.nan, 6, 3, 5.5, np.nan, 8, 4, 3.5, 2.5, 4.5, 3, 2.0, 5.0,
            6.0, 7.0, 2.0, 4.0, 3.0, 5.5, 4.5, 7],
    'visits': [2, 4, 3, 4, 3, 4, 2, 2, 3, 2, 2, 1, 3, 2, 4, 3, 2, 1, 5, 3, 4, 1, 2, 5],
    'priority': ['yes', 'yes', 'no', np.nan, 'no', 'no', 'no', 'yes', 'no', 'no', 'yes', 'no',
                'yes', 'no', 'no', 'yes', 'no', 'yes', 'yes', 'no', 'no', 'yes', 'no', 'yes']
}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't',
'u', 'v', 'w', 'x']

birds = pd.DataFrame(data = data)
birds['labels'] = labels
birds 
```

```python
birds.isnull().sum()
```

```python
ageMode = birds['age'].mode()[0]
priMode = birds['priority'].mode()[0]
```

```python
birds['age'] = birds['age'].fillna(ageMode)
birds['priority'] = birds['priority'].fillna(priMode)
birds
```

```python
# Inequalities should be individual / binary only

birds[(birds['birds'] == 'Cranes') & (2 <= birds['age']) & (birds['age'] <= 4)]
```

```python
birds.sort_values(by = ['age', 'visits'], ascending = [False, True])
```

```python
birds = birds.drop_duplicates()
```

```python
def yees(values):
    c = 0
    for data in values:
        if data == 'yes':
            c += 1
    return c

def noo(values):
    c = 0
    for data in values:
        if data == 'no':
            c += 1
    return c

pivot_df = birds.pivot_table(values = 'priority', index = 'birds', aggfunc = [yees, noo])
pivot_df.plot(kind='bar', figsize=(10, 6))
plt.title('Yes and no for each priority')
plt.xlabel('Bird type')
plt.ylabel('Priority')
plt.xticks(rotation = 0)  # Rotate x labels for better readability
plt.legend(title='Statistics')
plt.show()
```
