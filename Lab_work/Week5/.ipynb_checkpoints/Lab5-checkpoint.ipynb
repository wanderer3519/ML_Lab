```python
'''
1 Naive Bayes
1.1A. Load the given Customer Churn dataset (whether the customer will leave the service or
not) for the classification task
'''
```

```python
import pandas as pd
import numpy as np
import scipy as sp
import sklearn
```

```python
# Load the dataset
cc = pd.read_csv("./Cust_Churn.csv")
cc
```

```python
# 1.1B. Check for null values. Drop the unnecessary column.
cc['Churn']
df = cc.drop(columns = ['CustomerID', 'PaperlessBilling', 'PaymentMethod'])
df
```

```python
# Check for null values

df.isnull().sum()
```

```python
'''
1.1C. Find out the columns with the categorical values and convert it into numerical data using
appropriate technique
'''

categ = set()
for cat, val in df.items():
    if val.dtype == 'object':
        categ.add(cat)
categ
```

```python
# Label encoding
from sklearn.preprocessing import LabelEncoder, StandardScaler

newM = LabelEncoder()
catDict = {i: LabelEncoder() for i in categ}

for cat in catDict:
    catDict[cat].fit(df[cat])
    df[cat] = catDict[cat].transform(df[cat])
df
```

```python
'''
1.1D. Find out the columns with the numerical values and apply the StandardScaler to the
numerical columns.
'''
stdsc = StandardScaler()
data = np.array([])
for label in df:
    if label not in categ and label != 'Churn':
        if not len(data):
            data = np.array(df[label])
        data = np.column_stack([data, df[label]])

data
```

```python
stdsc.fit(data)
```

```python
x = stdsc.transform(data)
x
```

```python
features = x
target = np.array(df['Churn'])
```

```python
# 1.1E. Separate features and target column(Churn).
from collections import Counter
count = Counter(target) 
count
```

```python
'''
    1.1F Check the class distribution of the target variable (Churn). Write your observation.
    Observation for class distribution
    The # zeros is more than # ones

    Mean -> 0.18
    Median -> 0
    Mode -> 0
'''
```

```python
# 1.1G. Split the data into train-tests in a ratio 8:2
from sklearn.model_selection import train_test_split as TTS

xTrain, xTest, yTrain, yTest = TTS(features, target, test_size = 0.20, random_state = 42)
```

```python
xTrain.shape, yTrain.shape
```

```python
yTrain
```

```python
# 1.2A Init SMOTE and 1.2B apply SMOTE to train data
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state = 42)
sm.fit(xTrain, yTrain)
```

```python
# Resample the data
xTrain2, yTrain2 = sm.fit_resample(xTrain, yTrain)
Counter(yTrain2)

# 1.2C Class Distribution after SMOTE
# Observation == Ytrain has the same number of 1s and 0s
```

```python
'''
    1.3A. Initialize Naive Bayes with specified hyperparameter var_smoothing=1e-09
    The values given here are the default hyper parameters mentioned in doc. So not initializing it
'''
from sklearn.naive_bayes import GaussianNB


# 1.3B. Train the Naive Bayes model on the resampled training data
gnb = GaussianNB()
gnb.fit(xTrain2, yTrain2)
```

```python
# 1.3C predict the test data

yPred = gnb.predict(xTest)
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

# 1.3D Mention the metrics

cfmat = confusion_matrix(yTest, yPred)
accScore = accuracy_score(yTest, yPred)
cfrep = classification_report(yTest, yPred) 
print(cfrep)

# 1.3E Write your observation about the performance of your model. Which metric are you considering and why?
# Recall is better metric because there is much less spread in  its values
```

```python
# 2.1A Init Logistic regression
from sklearn.linear_model import LogisticRegression

# 2.1B Training on resampled data
lr = LogisticRegression(C = 0.01, max_iter = 100, penalty = 'l2', solver = 'liblinear')
lr.fit(xTrain2, yTrain2)
```

```python
# 2.1C Predeict test
# 2.1D Evaluate model using metrics

yPredLR = lr.predict(xTest)
cfmatLR = confusion_matrix(yTest, yPredLR)
accScreLR = accuracy_score(yTest, yPred)
cfrepLR = classification_report(yTest, yPred)

print(cfrepLR)
```

```python
# Observation: More improved than the earlier ones. I consider the recall metric as it is more stable across 0 and 1
```

```python
# 2.2A Predict probabilities for Logistic Regression

probLR = lr.predict_proba(xTest)
yScoreLR = probLR[:, 1]
```

```python
from sklearn.metrics import roc_auc_score, roc_curve

# 2.2B compute metrics
rocAucScore = roc_auc_score(yTest, yScoreLR)
rocAucScore
```

```python
# 2.2C predict probability for Naive Bayes

probNB = gnb.predict_proba(xTest)
yScoteNB = probNB[:, 1]

rocAucScore = roc_auc_score(yTest, yScoteNB)
rocAucScore
```

```python
# 2.2D Compute metrics
NBfpr, NBtpr, tNB = roc_curve(yTest, yScoteNB)
LRfpr, LRtpr, tLR = roc_curve(yTest, yScoreLR)
```

```python
from matplotlib import pyplot as plt

plt.title("ROC curves of different classifiers")
plt.xlabel("False positive rate")
plt.ylabel("True positive rate")

plt.plot(NBfpr, NBtpr, label = "Logistic Regression")
plt.plot(LRfpr, LRtpr, label = "Naive Bayes")

plt.legend()

# Logistic regression does better
```
